```python
from os import path

import sys
sys.path.append(path.join("..", "notebooks"))
sys.path.append(path.join("..", "notebooks", "utils.py"))
import utils

import anndata
import numpy as np
import pandas as pd
import scipy.sparse as sp
import string

import matplotlib.pyplot as plt
```

# Baron

```python
adata_250 = anndata.read_h5ad("baron_embedding_tsne_250_genes.h5ad")
adata_3000 = anndata.read_h5ad("baron_embedding_tsne_3000_genes.h5ad")
adata_full = anndata.read_h5ad("baron_embedding_tsne_all_genes.h5ad")
```

```python
print(adata_full.shape[1], "genes")
```

```python
colors = utils.get_colors_for(adata_3000)
```

```python
cell_order = list(colors.keys())
num_cell_types = len(np.unique(adata_250.obs["labels"]))

fig, ax = plt.subplots(ncols=3, figsize=(15, 5))

utils.plot(adata_250.obsm["tsne"], adata_250.obs["labels"],  s=1, colors=colors, draw_legend=False, ax=ax[0], alpha=1, title="250 genes", label_order=cell_order)
utils.plot(adata_3000.obsm["tsne"], adata_250.obs["labels"], s=1, colors=colors, draw_legend=False, ax=ax[1], alpha=1, title="3,000 genes", label_order=cell_order)
utils.plot(adata_full.obsm["tsne"], adata_250.obs["labels"], s=1, colors=colors, draw_legend=True, ax=ax[2], alpha=1, title="17,499 genes", label_order=cell_order,
          legend_kwargs=dict(loc="upper center", bbox_to_anchor=(0.5, 0.05), bbox_transform=fig.transFigure, labelspacing=1, ncol=num_cell_types // 2 + 1))

for ax_, letter in zip(ax, string.ascii_lowercase): 
    plt.text(0, 1.02, letter, transform=ax_.transAxes, fontsize=15, fontweight="bold")

for ax_ in ax: ax_.axis("equal")

plt.savefig("baron_embedding_tsne_genes.pdf", dpi=600, bbox_inches="tight", transparent=True)
```
